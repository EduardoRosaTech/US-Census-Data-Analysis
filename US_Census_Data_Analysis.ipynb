```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (adjust path if necessary)
df = pd.read_csv('/content/us_census.csv')

# Strip column names and check data
df.columns = df.columns.str.strip()
print(df.info())
print(df.describe())

#  Plot 1: Average income by state
plt.figure(figsize=(10,6))
sns.barplot(
    x='State',
    y='Income',
    data=df.sort_values('Income', ascending=False),
    palette='viridis',
    hue='State',      # Required to use palette
    legend=False      # Avoid duplicate legend
)
plt.title('Average Income by State (USD)')
plt.xlabel('State')
plt.ylabel('Average Income (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#  Plot 2: Correlation heatmap between key variables
plt.figure(figsize=(8,5))
sns.heatmap(
    df[['Income', 'Poverty', 'Bachelor_or_Higher']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Matrix: Income, Poverty, Education')
plt.tight_layout()
plt.show()

```
