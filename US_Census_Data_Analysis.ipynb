```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('/content/us_census.csv')

# Preview
df.head()

# Clean column names
df.columns = df.columns.str.strip()

# Basic info
print(df.info())
print(df.describe())

# Plot: Average income per state
plt.figure(figsize=(10,6))
sns.barplot(x='State', y='Income', data=df.sort_values('Income', ascending=False), palette='viridis')
plt.title('Average Income by State (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df[['Income', 'Poverty', 'Bachelor_or_Higher']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
